```python
#Top 10 Highest Growing Countries (2022)
import matplotlib.pyplot as plt
import seaborn as sns

# Select the top 10 fastest-growing economies in 2022
Top_22 = data[['Country Name', '2022 [2022]']].sort_values(by='2022 [2022]', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x='2022 [2022]', y='Country Name',hue='Country Name', data=Top_22, palette='viridis',legend=False)

# Labels and Title
plt.xlabel('GDP Growth (%)')
plt.ylabel('Country')
plt.title('Top 10 Highest Growing Countries in 2022')
plt.xlim(0, Top_22['2022 [2022]'].max() + 1)  # Adjust x-axis for better visualization
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python

data['2025 [2025]'] = pd.to_numeric(data['2025 [2025]'], errors='coerce')

# Select the top 10 fastest-growing economies in 2025
Top_22 = data[['Country Name', '2025 [2025]']].sort_values(by='2025 [2025]', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x='2025 [2025]', y='Country Name', hue='Country Name', data=Top_22, palette='viridis', legend=False)

# Labels and Title
plt.xlabel('GDP Growth (%)')
plt.ylabel('Country')
plt.title('Top 10 Highest Growing Countries in 2025')  # Corrected title year
plt.xlim(0, Top_22['2025 [2025]'].max() + 1)  # Adjust x-axis for better visualization
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python

data['2026 [2026]'] = pd.to_numeric(data['2026 [2026]'], errors='coerce')

# Select the top 10 fastest-growing economies in 2025
Top_22 = data[['Country Name', '2026 [2026]']].sort_values(by='2026 [2026]', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x='2026 [2026]', y='Country Name', hue='Country Name', data=Top_22, palette='viridis', legend=False)

# Labels and Title
plt.xlabel('GDP Growth (%)')
plt.ylabel('Country')
plt.title('Top 10 Highest Growing Countries in 2026')  # Corrected title year
plt.xlim(0, Top_22['2026 [2026]'].max() + 1)  # Adjust x-axis for better visualization
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python

# Plot the GDP Growth Trend from 2022 to 2026 for Top 10 Countries

Top_10_2022 = data[['Country Name', '2022 [2022]', '2023 [2023]', '2024 [2024]', '2025 [2025]', '2026 [2026]']].sort_values(by='2022 [2022]', ascending=False).head(10)

plt.figure(figsize=(14, 8))

# Loop through the top 10 countries to plot their GDP growth trends
for country in Top_10_2022['Country Name']:
    country_data = Top_10_2022[Top_10_2022['Country Name'] == country]
    plt.plot(['2022', '2023', '2024', '2025', '2026'], country_data.iloc[:, 1:].values.flatten(), label=country, marker='o')

plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')
plt.title('Top 10 Fastest-Growing Countries: GDP Growth Trend from 2022 to 2026')
plt.legend(title="Countries", bbox_to_anchor=(1.05, 1), loc='upper left')  
plt.grid(True)

# Show plot
plt.tight_layout()  
plt.show()
```
