```python
# Importing Libraries
```

```python
import numpy as np
import pandas as pd
```

```python
data =pd.read_csv('GDP Overview.csv')
data
```

```python
missing_data = data.isnull().sum()
print(missing_data)
#There is no missing data in the dataset
```

```python
# Descriptive Statistics
```

```python
gdp_stat = data.describe()
gdp_stat
```

```python
#Top 10 Highest Growing Countries (2022)
import matplotlib.pyplot as plt
import seaborn as sns

# Select the top 10 fastest-growing economies in 2022
Top_22 = data[['Country Name', '2022 [2022]']].sort_values(by='2022 [2022]', ascending=False).head(10)

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x='2022 [2022]', y='Country Name',hue='Country Name', data=Top_22, palette='viridis',legend=False)

# Labels and Title
plt.xlabel('GDP Growth (%)')
plt.ylabel('Country')
plt.title('Top 10 Highest Growing Countries in 2022')
plt.xlim(0, Top_22['2022 [2022]'].max() + 1)  # Adjust x-axis for better visualization
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```
